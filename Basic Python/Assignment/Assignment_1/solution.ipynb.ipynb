# Assignment 1: Temperature Check 
Problem: 
 Write a program that takes the current temperature as input and prints: <br>
●  "It's too cold!" if below 10°C  
●  "It's cool outside" if between 10°C and 25°C  
●  "It's hot outside!" if above 25°C

```python
temp=int(input("Enter your current tempareture: "))
if temp<10:
    print("It's too cold!")
elif(10<=temp<=25): # temp>=10 and temp <=25, we can write this logic like this
    print("It's cool outside")
else:
    print("It's hot outside!")
```

# Assignment 2: ATM Withdrawal System 
Problem: 
 Ask the user for their account balance and the amount they want to withdraw. <br>
●  If the withdrawal amount is greater than the balance → print "Insufficient 
Balance"  <br>
●  If the withdrawal amount is less than or equal to balance → print "Transaction 
Successful" and show the remaining balance. <br>

```python
balance=5000
decision=int(input("""Choose what you want
          1. Check balance
          2. Withdrawal balance
          
          """))
if(decision==1):
    print(f"Your balance is: {balance}")
elif(decision==2):
    input_balance=int(input("Enter your withdrawl amount that you want to withdraw: "))
    if(input_balance>balance):
        print(f"Insufficient. Your current balance is: {balance}")
    else:
        balance-=input_balance
        print("Transaction Successful")
        print(f"Your current balance is: {balance}")
else:
    print("Press the right key")
```

# Assignment 3: Student Grade Calculator 
Problem: 
 Input marks (0–100) and print grade based on: <br>
●  90–100 → A+  
●  80–89 → A  
●  70–79 → B  
●  60–69 → C  
●  Below 60 → Fail  
 


```python
marks=int(input("Enter your marks(0-100): "))

if(0<=marks<=100):
    if(90<=marks<=100):
        print("A+")
    elif(80<=marks<=89):
        print("A")
    elif(70<=marks<=79):
        print("B")
    elif(60<=marks<=69):
        print("C")
    elif(marks<60):
        print("Fail")
else:
    print("Enter number between 0 to 100")
```

# Assignment 4: Bus Ticket Discount 
Problem: 
 Ask the user for age and print ticket price: <br>
●  Below 5 → Free  
●  5–18 → 50% Discount  
●  60 or above → 30% Discount  
●  Others → Full Price

```python
ticket_price=2000
age=int(input("Enter your age: "))
if(age<5):
    print("Free")
elif(5<=age<=18):
    ticket_price*=0.5
    print(f"50% Discount. Ticket price is {int(ticket_price)}")
elif(age>=60):
    ticket_price*=0.7
    print(f"30% Discount. Ticket price is {int(ticket_price)}")
else:
    print(f"Full price. Ticket price is {ticket_price}")
```

# Assignment 5: Restaurant Bill with Discount 
Problem: 
 Take total bill amount from the user: <br>
●  If the bill is above 1000 → apply 10% discount  
●  Otherwise → no discount  Then print the final bill amount.

# Assignment 8: Leap Year Checker 
Problem: 
 Take a year as input and check: <br>
●  If it is divisible by 400 → Leap Year  
●  Else if divisible by 4 but not by 100 → Leap Year  
●  Otherwise → Not a Leap Year

```python
year=int(input("Enter a year: "))
if(year%400==0):
    print("Leap year!")
elif(year%4==0 and year%100!=0):
    print("Leap year!")
else:
    print("Not leap year!")
```
