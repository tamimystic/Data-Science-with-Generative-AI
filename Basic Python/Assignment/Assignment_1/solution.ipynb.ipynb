# Assignment 1: Temperature Check 
Problem: 
 Write a program that takes the current temperature as input and prints: <br>
●  "It's too cold!" if below 10°C  
●  "It's cool outside" if between 10°C and 25°C  
●  "It's hot outside!" if above 25°C

```python
temp=int(input("Enter your current tempareture: "))
if temp<10:
    print("It's too cold!")
elif(10<=temp<=25): # temp>=10 and temp <=25, we can write this logic like this
    print("It's cool outside")
else:
    print("It's hot outside!")
```

# Assignment 2: ATM Withdrawal System 
Problem: 
 Ask the user for their account balance and the amount they want to withdraw. <br>
●  If the withdrawal amount is greater than the balance → print "Insufficient 
Balance"  <br>
●  If the withdrawal amount is less than or equal to balance → print "Transaction 
Successful" and show the remaining balance. <br>

```python
balance=5000
decision=int(input("""Choose what you want
          1. Check balance
          2. Withdrawal balance
          
          """))
if(decision==1):
    print(f"Your balance is: {balance}")
elif(decision==2):
    input_balance=int(input("Enter your withdrawl amount that you want to withdraw: "))
    if(input_balance>balance):
        print(f"Insufficient. Your current balance is: {balance}")
    else:
        balance-=input_balance
        print("Transaction Successful")
        print(f"Your current balance is: {balance}")
else:
    print("Press the right key")
```

# Assignment 3: Student Grade Calculator 
Problem: 
 Input marks (0–100) and print grade based on: <br>
●  90–100 → A+  
●  80–89 → A  
●  70–79 → B  
●  60–69 → C  
●  Below 60 → Fail  
 


```python
marks=int(input("Enter your marks(0-100): "))

if(0<=marks<=100):
    if(90<=marks<=100):
        print("A+")
    elif(80<=marks<=89):
        print("A")
    elif(70<=marks<=79):
        print("B")
    elif(60<=marks<=69):
        print("C")
    elif(marks<60):
        print("Fail")
else:
    print("Enter number between 0 to 100")
```
