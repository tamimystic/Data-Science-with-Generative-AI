# Loop Control Statements


### 3 types of loop control statements
1. Break
2. Continue
3. Pass

## 1. Break

```python
#if we need to stop code
for i in range (1,11):
    if (i==5):
        break
    print(i)
```

```python
names=["joe","doe","hue","Tamim","alex"]

for i in names:
    if (i=="Tamim"):
        print(i)
        break
    print(names,"------")
```

## 2. Continue

```python
#one kind of skip
for i in range(1,10):
    if i==5:
        continue
    print(i)
```

## 3. Pass

```python
# if we don't know what is the logic or what will be our code
for i in range(1,11):
    if(i==5):
        pass
    print(i)
```
