* if i want to print 'Tamim' 20 times, it is need to write 20 times manually, but this is not efficient, for this case we can use while loop

```python
# While Loop
i=0

while i<10: # it will run untill it (i<10) is false
    print("Tamim Hossain")
    i+=1
```

childhood multiplication using while loop

```python
num=int(input("Enter your number: "))
i=1
while(i<=10):
    print(i*num)
    i+=1
```

```python
#human readable multiplication

num=int(input("Enter your number: "))
i=1
while(i<=10):
    print(num, " * ", i," = ", num*i)
    i+=1
```
