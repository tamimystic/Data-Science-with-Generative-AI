Make a game using while loop and random funtion

```python
#simple logic
import random
randomNumber=random.randint(a=1, b=50)
userNumber=int(input("Enter a number (1-50): "))
while randomNumber!=userNumber:
    userNumber=int(input("Enter a number (1-50): "))
    if(userNumber<randomNumber):
        print("Enter bigger number!")
    elif(userNumber>randomNumber):
        print("Enter lower number!")
    else:
        print("Wow! you got it. Number is: ",userNumber)
        
```

```python
# Count attempt time
import random
randomNumber=random.randint(a=1, b=50)

counter=0;
while True:
    userNumber=int(input("Enter a number (1-50): "))
    counter+=1
    if(userNumber<randomNumber):
        print("Enter bigger number!")
    elif(userNumber>randomNumber):
        print("Enter lower number!")
    else:
        print("Wow! you got it. Number is: ",userNumber)
        print("Your attempt is: ", counter)
        break
```
