# String Operation

* Arithmatic operation
* Relational operation
* Logical operation
* Loops in strings
* Membership operation

### 1. Arithmatic operation

```python
print("Tamim" + " " + "Hossain")
```

```python
print("Tamim "*100)
```

### 2.Relational operation in strings

```python
print("tamim"=="tamim") #equality
print("Tamim"!="Hossain") # not equal
print("Hello">"Hi") #grater than (order by oxford dictionary)
print("Hello"<"Hi") #less than (order by oxford dictionary)
print("Hello"<="Hello") #grater than and equal
```

### 3.Logical operation in strings

and

```python
"hello" and "world"
```

```python
"tamim" and ""
```

or

```python
"tamim" or "hossain"
```

```python
" " or "hossain"
```

not

```python
not "tamim"
```

### 4. Loops in strings

```python
for i in "Tamim":
    print(i)
```

### 5. Membership operation

```python
"T" in "Tamim Hossain"
```

```python
"U" in "Tamim Hossain"
```
