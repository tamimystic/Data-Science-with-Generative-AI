```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv(r"C:\Inception BD\Data-Science-with-Generative-AI\AI final lab test (VU)\sports_data.csv")

print("Shape:", data.shape)
print("\nHead:\n", data.head())
print("\nInfo:")
print(data.info())
print("\nDescribe:\n", data.describe())

print("\nMissing Values:\n", data.isnull().sum())

plt.figure(figsize=(6,4))
sns.countplot(data=data, x="class")
plt.title("Class Distribution")
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(data["height"], kde=True)
plt.title("Height Distribution")
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(data["weight"], kde=True)
plt.title("Weight Distribution")
plt.show()

plt.figure(figsize=(7,5))
sns.scatterplot(data=data, x="height", y="weight", hue="class", s=80)
plt.title("Height vs Weight by Class")
plt.show()

plt.figure(figsize=(5,4))
sns.heatmap(data[["height","weight"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

# Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(data=data[["height","weight"]])
plt.title("Boxplot of Height & Weight")
plt.show()
```
