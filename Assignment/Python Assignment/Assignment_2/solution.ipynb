3. Password Strength Checker  
�� Real-world use: Used in user authentication systems.  
Task:  
Check if a password is strong — must have at least one uppercase, one lowercase, one digit, and one  special character.  
Example:  
Input: "Hello@123"  
Output: Strong Password 
 


```python
pas=input("Enter your password: ")
```

```python
pas
```

```python
upper=False
lower=False
digit=False
others=False

for i in pas:
    if 'A'<=i<='Z':
        upper=True
    elif 'a'<=i<='z':
        lower=True
    elif '0'<=i<='9':
        digit=True
    else:
        others=True
```

```python
if upper and lower and digit and others:
    print("Strong Password")
else:
    print("not strong")
```

```python
text
```

9. Count Hashtags and Mentions (Social Media 
Analytics) �� Real-world use: Used in analyzing social media posts.  
Task:  
Count how many hashtags (#) and mentions (@) appear in a post.  
Example:  
Input: "Check out #AIWorkshop and follow @inceptionbd"  
Output: Hashtags = 1, Mentions = 1 

```python
text=input("Enter your text: ")
```

```python
hashCounter=0
mentionCounter=0
for i in text:
    if i=='#':
        hashCounter+=1
    elif i=='@':
        mentionCounter+=1
    else:
        pass
print(f"Hashtags = {hashCounter}, Mentions = {mentionCounter}")
```

 
10. Word Reversal for Data Encryption  
�� Real-world use: Simple text obfuscation or playful encoding.  
Task:  
Reverse each word in a sentence while keeping word order the same.  
Example:  
Input: "AI is fun"  
Output: "IA si nuf" 

```python
takeInput=input("Enter your text: ").split()
```

```python
takeInput
```

```python
takeInput[::-1]
```

```python
rev=[]
for i in takeInput:
    rev.append(i[::-1])
    print(rev)
print(" ".join(rev))
```
