1. Email Validator (String Filtering + Loops)
�� Real-world use: Used in signup forms or data cleaning tasks. <br>
Task:
Check if a given email is valid — must contain “@” and end with “.com” or “.org”.
Example:
Input: hello@inceptionbd.com
Output: Valid Email


```python
email=input("Enter your email: ")
```

```python
email[-4:]
```

```python
for i in email:
    if i=='@' and (email[-4:]=='.com' or email[-4:]=='.org') :
        print('Valid Email')
        break
else:
    print("Please Enter Valid Email")
```

2. Count Word Frequency in Customer Reviews
�� Real-world use: Analyzing reviews or comments in text analytics.
Task:
Given a review text, count how many times each word appears (ignore case).
Example:
Input: "This product is good and this service is excellent"
Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
t=input("enter your text: ")
l=t.lower()
```

```python
l
```

```python
s=l.split()
```

```python
s
```

```python
s.count("is")
```

```python
dekha=[]
for i in s:
    if i not in dekha:
        print(i,"=",s.count(i), end=", ")
        dekha.append(i)
```

3. Password Strength Checker  
�� Real-world use: Used in user authentication systems.  
Task:  
Check if a password is strong — must have at least one uppercase, one lowercase, one digit, and one  special character.  
Example:  
Input: "Hello@123"  
Output: Strong Password 
 


```python
pas=input("Enter your password: ")
```

```python
pas
```

```python
upper=False
lower=False
digit=False
others=False

for i in pas:
    if 'A'<=i<='Z':
        upper=True
    elif 'a'<=i<='z':
        lower=True
    elif '0'<=i<='9':
        digit=True
    else:
        others=True
```

```python
if upper and lower and digit and others:
    print("Strong Password")
else:
    print("not strong")
```
