* 1. Email Validator (String Filtering + Loops)
�� Real-world use: Used in signup forms or data cleaning tasks. <br>
Task:
Check if a given email is valid — must contain “@” and end with “.com” or “.org”.
Example:
Input: hello@inceptionbd.com
Output: Valid Email


```python
email=input("Enter your email: ")
```

```python
email[-4:]
```

```python
for i in email:
    if i=='@' and (email[-4:]=='.com' or email[-4:]=='.org') :
        print('Valid Email')
        break
else:
    print("Please Enter Valid Email")
```
