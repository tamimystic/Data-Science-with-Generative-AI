## Extend

```python
l=[1,2,3,4]
a=[10,11,12]

l.extend(a)
l
```

## Insert
### add value in specific position

```python
l=[1,2,3,4]
l.insert(1,200) # here 1 is position and 200 is value
print(l)
```

## Delete and Remove
* del -> index define
* remove -> value define 

```python
l=[1,2,3,4]
del l[1:2] # del 
l
```

```python
l=[1,2,3,4]
l.remove(3)
l
```
