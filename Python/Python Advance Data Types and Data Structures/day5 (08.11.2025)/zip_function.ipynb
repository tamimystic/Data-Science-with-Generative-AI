The zip() function in Python is a built-in function that combines multiple iterables (like lists, tuples, or strings) into a single iterator of tuples. Each tuple contains elements from the input iterables that are at the same index.

```python
l1=[1,2,3,4]
l2=[6,7,8,9]
```

```python
list(zip(l1,l2)) #zip gives us tupple
```

```python
l3=[11,22,33]
list(zip(l1,l2,l3))
```

```python
roll=[1,2,3,4]
name=["tamim","hossain"]

list(zip(roll,name))
```

```python
for i in zip(l1,l2):
    print(i)
```
