Provides a concise way of creating list

```python
# add 1 to 10 number a list
l=[]
for i in range(1,11):
    print(i)
    l.append(i)
else:
    print[l]
```

```python
lc=[i for i in range (1,11)] #list comprehension
print(lc)
```

```python
# scalar multiplication on a vector
v=[2,3,4]
s=-3
l=[]
for i in v:
    result=i*s
    l.append(result)
print(l)
```

```python
lc=[i*s for i in v]
print(lc)
```

```python
for i in range(1,51):
    if i%5==0:
        print(i, "divisible")
```

```python
lc=[i for i in range(1,51) if i%5==0]
print(lc)
```

```python
# find language which starts with p
languages = ['java', 'python']
lc = [lang for lang in languages if lang.startswith('p')]
print(lc)
```
