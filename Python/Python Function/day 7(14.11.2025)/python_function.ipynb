# *args and **kwargs

*args and **kwargs are special Python keywords that are used to pass the variable length of arguments to a function

*args

allows us to pass a variable number of non-keyword arguments to a function.

## *args

if i need 2 parameter

```python
def multiply(a,b):
    return a*b
```

```python
res=multiply(2,3)
print(res)
```

if i need 3 parameter

```python
def multiply(a,b,c):
    return a*b*c
```

```python
res=multiply(2,3,4)
print(res)
```

if i need 4 parameter

```python
def multiply(a,b,c,d):
    return a*b*c*d
```

```python
res=multiply(2,3,4,5)
print(res)
```

```python
def multiply(*args):
    print(args)
```

```python
multiply(1,2,3,45,5,7)
```

```python
products=1
def gun(*args):
    global products
    for i in args:
        products*=i
    return products
d=gun(1,2,3,4,5)
print(d)
```

## **kwargs
-> it is for key-value pair

```python
def displays(**kwargs):
    print(kwargs)

displays(name="Tamim", id=223311232)
```

```python
def displays(**kwargs):
    for i,j in kwargs.items():
        print(i,"->",j)

displays(name="Tamim", id=223311232)
```

### Variable Scope in Functions
1. global variables
2. local variables

```python
def g(y):
    print(x)
    print(x+1) # it does not impact in actual variable

x=5 #global variable
g(x)
print(x)
```

```python
def f(y):
    x=1 #local (here x can be 5, but x takes local value)
    x+=1
    print(x)
x=5 #global
f(x)
print(x)
```
