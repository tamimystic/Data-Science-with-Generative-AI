A function in another function

```python
def f():
    def g():
        print("G Function")
    print("F Function")
    
```

```python
f()
```

```python
def f():
    def g():
        print("G Function")
        
    g()

    print("F Function")

f()
```

```python
def f():
    def g():
        print("G Function")
        f()
        
    g()

    print("F Function")

f()
```

## Returning a function
