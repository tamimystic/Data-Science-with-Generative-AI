# Enumerate Function in python

```python
L = ["Tamim", "Alex", "Hakim", "Dummy"]

for name in L:
    print(name)
```

```python
l = ["Tamim", "Alex", "Hakim", "Dummy"]

for index,name in enumerate(l,start=1):
        print(index,name)
```

```python
l = ["Tamim", "Alex", "Hakim", "Dummy"]

for index,name in enumerate(l,start=1):
    if index==3:
        print(index,name)
```
