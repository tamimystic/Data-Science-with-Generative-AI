#### Nested Loop

```python
for i in range (1,6):
    for j in range (1,6):
        print(i,j)
```

```python
row=int(input("Enter your rows number: "))
for i in range(1,row+1):
    for j in range(1,i+1):
        print("*", end="")
    print("")
```
