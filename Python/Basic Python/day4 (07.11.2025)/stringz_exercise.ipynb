3. write a program that can check whether a given string is palindrome or not

- abba
- mam
- madam

```python
s=input("Enter your word: ")

if(s==s[::-1]):
    print("palindrome")
else:
    print("not palindrome")
```

```python
#Another approach
word=input("Enter your word: ")

flag=True
for i in range(0,len(word)//2):
    if word[i] != word[len(word)-i-1]:
        flag=False
        print("Not palindrome")
        break
if flag:
    print("Palindrome")
```
