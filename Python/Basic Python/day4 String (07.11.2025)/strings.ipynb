#### access multiple character/slice character, that means if i access "Tamim"

```python
s[0:5] #here need to put this strings 1st index and last index, and here it will run 1st<last
```

```python
s[6:13] #here 'H' is 1st index and last 'n' index +1
```

```python
print(s[6:13])
```

```python
s[6: 500] #if put last number as my wish, it will be no probelm
```

```python
s[6:] #if do not put last number, it will be no probelm
```

```python
s[:] #whole string
```

```python
print(s[0:5:2]) #0 is 1st index, 5 is last index and 2 is 'jump parameter', that means normally it print 1 by 1, now print 2 chracter then print next 2 number character
```

```python
s[::2] #print whole string with jump 2nd by 2nd character, that means skip 1 character
```

```python
print(s[4::-1]) #if i want to print a string in reverse
```

```python
print(s[::-1]) #print it normally reverse
```

```python
s[::-2]
```
