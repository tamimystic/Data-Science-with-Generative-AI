## Problems with working in text mode

- can't work with binary files like images  
- not good for other data types like int / float / list / tuples

- here in image data, read/write/append position we use read binary ("rb")

```python
print("image data")
```

```python
with open("tamim.jpg", "rb") as f:
    s=f.read()
    print(s)
```

* we can not see this photo

we can not see that photo, but we can make duplicate this photo without seen this

```python
with open("tamim.jpg", "rb") as f:
    original_file=f.read()
    with open("tamim_copy.jpg", "wb") as wf:
        wf.write(original_file)
```
