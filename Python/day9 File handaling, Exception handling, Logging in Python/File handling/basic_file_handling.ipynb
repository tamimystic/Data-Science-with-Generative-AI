## Types of data used for I/O

- **Text** – '12345' as a sequence of unicode chars  
- **Binary** – 12345 as a sequence of bytes of its binary equivalent  that means picture or something backend matrix value or something like that

## File Types

- **Text files** – All program files are text files  
- **Binary files** – Images, music, video, exe files

if we need read whole text using readline

```python
f=open("C:\Inception BD\Data-Science-with-Generative-AI\Python\day9 File handaling, Exception handling, Logging in Python\File handling\demo1.txt", "r")

while True:
    data=f.readline()

    if data== "":
        break
    else:
        print(data, end='')

f.close()
```
