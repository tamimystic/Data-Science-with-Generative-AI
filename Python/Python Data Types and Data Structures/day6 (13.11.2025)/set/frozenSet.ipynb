frozen set is just an immutable version of a python set object
* Immutability
* Unordered
* Unique Elements

```python
fs1=frozenset([1,2,3,4,5])
fs2=frozenset([3,4,5,6,7])
```

```python
fs=frozenset([1,2,3,4,5,(4,5)])
```

```python
fs
```

```python
my_list = [1, 2, 3, 2, 4]
frozen_set_from_list = frozenset(my_list)
print(frozen_set_from_list) 

my_tuple = ("apple", "banana", "cherry")
frozen_set_from_tuple = frozenset(my_tuple)
print(frozen_set_from_tuple) 

empty_frozenset = frozenset()
print(empty_frozenset)
```
