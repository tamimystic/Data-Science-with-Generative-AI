- int data can not be write/stored as int, it can be store as string

```python
with open("hello.txt", "w") as f:
    # data=f.write(5) , can't do this, but we can write it as string
    data=f.write("5")
    print(data)
```

* dictionary can not be write/stored as dictionary, it can be store as string

```python
d = {
    'name': 'Bappy',
    'age': 23,
    'gender': 'male'
}

with open('hello.txt', 'w') as f:
    f.write(str(d))
    f.write("\n\n\n")
```

```python
with open("hello.txt","r") as f:
    s=f.read()
    print(s)
    print(type(s))
```

```python
import json

# more complex data
d = {
    'name': 'Bappy',
    'age': 23,
    'gender': 'male'
}

with open('hello.txt', 'w') as f:
    json.dump(d, f, indent=4)
```

```python
with open("hello.txt","r") as f:
    s=f.read()
    print(s)
    print(type(s))
```
