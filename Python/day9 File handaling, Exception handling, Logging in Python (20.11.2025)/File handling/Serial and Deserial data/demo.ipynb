## Serialization and Deserialization

- **Serialization**  process of converting Python data types to JSON format  
- **Deserialization**  process of converting JSON to Python data types  

---

## What is JSON?

The full form of JSON is JavaScript Object Notation.



# List

```python
import json

l=[1,2,3,4]

with open("list.json","w") as f:
    json.dump(l,f)
```

```python
with open("list.json","r") as f:
    data=json.load(f)
    print(data)
    print(type(data))
```

# Dictionary

```python
dic={"tamim": {
        "name": "tamim",
        "id": 223311232,
    },
    "hakim": {
         "name":"hakim",
         "id": "223311201"
    }

}
with open("dict.json", "w") as f:
    json.dump(dic,f, indent=4)
```

```python
with open("dict.json", "r") as f:
    data=json.load(f)
    print(f)
    print(data)
    print(type(data))
    print(data["tamim"]["id"])
```

# Pickling in Python

## Pickling

**Pickling** is the process whereby a Python object hierarchy is converted into a byte stream,  
and **unpickling** is the inverse operation, whereby a byte stream (from a binary file or bytes-like object)  
is converted back into an object hierarchy.

### that means any object can store/retrive

```python
def display_function():
    return "This is Tamim"

d=display_function()
```

```python
d
```

```python
type(d)
```

```python
import pickle
with open("function.pkl","wb") as f:
    pickle.dump(d,f)
```

```python
with open("function.pkl","rb") as f:
    data=pickle.load(f)
    print(data)
    print(type(data))
```
