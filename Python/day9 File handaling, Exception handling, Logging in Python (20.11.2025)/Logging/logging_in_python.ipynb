```python
import logging
```

```python
import os
```

```python
log_dir="Logs"
log_file_name="application.log"
```

```python
os.makedirs(log_dir,exist_ok=True)
```

```python
log_path=os.path.join(log_dir,log_file_name)
```

```python
print(log_path)
```

```python
# for write string
logging.basicConfig(
    filename=log_path,
    format="[%(asctime)s] %(name)s - %(levelname)s - %(message)s",
    level=logging.INFO
    )
```

```python
def profile(name):
    # onek gula code

    if type(name)==str:
        print(f"wellcome {name}")
    else:
        logging.info("your input is not valid!")
```

```python
profile("Tamim")
```

```python
profile(8)
```

```python
def profile1(name):
    try:
        if type(name)==str:
            print(f"Hello {name}")
        else:
            raise TypeError("name must be string")
    except Exception as e:
        logging.info(str(e))
```

```python
profile1(786)
```

```python
def divide(a,b):
    try:
        if(a!=0):
            div=a/b
            logging.info(f"The divide result is: {div}")
        else:
            raise ZeroDivisionError("Can not divide number by zero")
    except Exception as e:
        logging.info(e)
```

```python
divide(8,7)
```
